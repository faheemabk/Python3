```python
i = 1
while i<=5:
    print(i)
    i = i+1
else:
    print("The else Suite")
print("Done with the loop")
#NOTE: The else block executes when the condition becomes false
```

```python
for i in range(6):
    print(i)
else:
    print("Done with the block")
print("Done with the Loop")
#Note: The else block is optional and is executds when the for loop is completed
```

```python
# The break statement is used inside a loop. It throws the control out of the loop when executed 
i = 0
while (i<=5):
    i = i+1
    print(i, end="  ")
print("Done with the loop")
```

```python
i = 0
while (i<=5):
    i = i+1
    if i==3:
        break
    print(i, end="  ")
    
print("Done with the loop")
```

```python
# The continue statement is used inside a loop
# It throws the control on the top of the loop when executed

i = 0
while (i<=5):
    i = i+1
    if i==3:
        
        print(i, end="  ")
else:
    print("The else of while")
print("Done with the loop")
```

```python
# The continue statement is used inside a loop
# It throws the control on the top of the loop when executed

i = 0
while (i<=5):
    i = i+1
    if i==3 or i==5:
        continue
    print(i, end="  ")
else:
    print("The else of while")
print("Done with the loop")
```

```python
# Nesting of loops
i = 1
while i<=3:
    j=1
    while j<=5:
        print(j, end='  ')
        j += 1
    i += 1
    print()
```

```python
# Nesting of loops
i = 1
while i<=3:
    j=1
    while j<=5:
        print(i, end='  ')
        j += 1
    i += 1
    print()
```

```python
# Using for loops instead of while loops
for i in range(5, 0, -1):          
    for j in range(5, 0, -1):          
        print(j, end='  ')
    print()
```
