```python
String: Strings in python are identified as contigous set of characters represented in the quotation marks
Python allows single('), double(") or triple(""" or ''') quotes to denote string literals, as long as the same type is used to start and end the string
Subset of string can be accessed using the slice operator([] and [:]) with indices starting from 0 in the beginning of the string and working their way from-1 to the end
The plus (+) sign is used to concatenate strings and* sign to repeat a string
```

```python
# Single quote string
s1= 'Welcome to python programming'
```

```python
print(s1)
```

```python
s2 = "I'm a programmer'"
```
