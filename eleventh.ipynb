```python
Dictionaries:
a Dictionary is a set of key value pair
They are similar to arrays but with user defined index
the keys are unique(within one dictionary)
A pair of curly brackets create an empty dictionary
dictionaries are ordered from python verysion 3.7 onwards. Before 3.7, dictionaries were unordered
```

```python
d = {"Name" : "Amir", "Age" : 7, 'Class' : 'First'}
print(d)
print(type(d))
```

```python
print(d["Name"])
```

```python
#if we attempt to access data item with a key which is not a part of the dictionary then an error is thrown
print(d["Percentage"])
```

```python
d['Percentage'] = 64.78
print(d)
```

```python
#updating existing keys
d['Age'] = 9
print(d)
```

```python
print(d)
del d['Percentage']
print(d)
```

```python
#removing a value from dicionary
d = {'Name': 'Amir', 'Age': 9, 'Class': 'First', 'age': 9}
d.clear()
print(d)
```

```python
d = {'Name': 'Amir', 'Age': 9, 'Class': 'First', 'age': 9}
d.clear()
print(d)
del d
print(d)
```

```python
# Properties of a dictionary keys : no duplicate keys are allowed. when duplicate keys are encountered, the last assignment wins
d = {'Name': 'Amir', 'Age': 9, 'Class': 'First', 'Name': 'Faheem'}

print(d)
```

```python
# We can have duplicate values
d = {'Company' : 'apple', 'Fruit':'Apple'}
print(d)
```

```python
d = {0:'a', 1:'b', 2:'c', 3:'d'}
print(d)
```

```python
# can traverse using a loop
for i in range(4):
    print(d[i])
```

```python
# Dictionary keys can be of any data type 
d = {42:'aaa', 2.45:'bbb', True:'ccc'}
print(d)
print(d[42])
print(d[2.45])
print(d[True])
```

```python
#Dictionary keys can be of any built in objects like data types and function
d = {int:1, float:2, bool:3}
print(d)
```

```python
d = {hex:1, bin:2, oct:3}
print(d)
```

```python
# Different ways for defining a dictionary

# Method 1: Using curly braces {}
a = {'Amir': 85, 'Golu': 85, 'Monu': 82}
print(a)

# Method 2: Using dict() with keyword arguments
b = dict(Amir=85, Golu=89, Monu=82)
print(b)

# Method 3: Using dict() with a list of tuples
c = dict([('Amir', 85), ('Golu', 89), ('Monu', 82)])
print(c)
```

```python
#Dictionary Methods
#get(key, data) -> returns the value of a given key if key is present else returns the second argument
d = {'Amir' : 85, 'Golu': 89, 'Monu': 82}
b = d.get('Amir', 100)
print(b)
b = d.get('Molu', 250)
print(b)
```

```python
a = {'Amir':85, 'Golu':89, 'Monu':82}
b = a.keys() # to get all keys
print(b)
print(list(b))
b = a.values()
print(b)
print(list(b))
```

```python
a = {'Amir':85, 'Golu':89, 'Monu':82}
b = a
print(a)
print(b)
```

```python
a['Sonu'] = 96
print(a)
print(b)
```

```python
#copy() is used to copy a dictionary
a = {'Amir':85, 'Golu': 89, 'Monu':82}
b = a.copy()
print(a)
print(b)
a['Sonu'] = 96
print(a)
print(b)
```

```python
#fromkeys() is used to create a new dictionary from keys of a sequence and values set to value
s = ('name', 'age', 'gender')
d = d.fromkeys(s)
print(d)
```

```python
s = ('name', 'age', 'gender')
d = d.fromkeys(s, 101)
print(d)
```

```python
#setdefault() -> similar to get() but will set dic[key]=value if the key is not present in the dictionary
d = {'Amir' : 85, 'Golu': 89, 'Monu': 82}
print(d)
print(d.setdefault('Golu', 111))
print(d)
```

```python
# items() => returns the list of directories(key, value) tuple pair
a = {'Amir' : 85, 'Golu': 89, 'Monu': 82}
print(a)
print(a.items())
print(list(a.items()))
```

```python
# update() -> merges a dictionary with another dictionary
a1 = {'Name':'amir', 'Age':7, "Class":"First"}
a2 = {'Amir': 87, 'salman':85, 'saif': 82, 'Age':23}
print(a1)
print(a2)
a1.update(a2)
print(a1)
print(a2)
```

```python
#pop(key, default) -> removes a key and returns its value if present else returns the second argument if 
#provided, else an error
d = {'a':10, 'b':20, 'c':30}
print(d)
x = d.pop('b')
print(x)
print(d)
```

```python
d = {'a':10, 'b':20, 'c':30}
print(d)
x = d.pop('e', -1)
print(x)
```

```python
#popitem() -> removes the LAST key-value pair from dictionary and returns as a tuple
d = {'a':10, 'b':20, 'c':30}
print(d)
x = d.popitem()
print(x)
print(d)
```

```python
#loop through a dictionary
d = {'Amir':87, 'Salman':85, 'Saif':82, 'Age':23}
#print all key names in a dictionary one by one
for x in d:
    print(x)
```

```python
#print all key names using the function keys()
d = {'Amir':87, 'Salman':85, 'Saif':82, 'Age':23}
#print all key names in a dictionary one by one
for x in d:
    print(x)
```
