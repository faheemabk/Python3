```python
Module:
A module allows us to logically organies the python code
It groups related code thus making the easier to handle and use
It is a file containing python code
A module can contain functions, classes, variables, runnable code, etc.
Modules enale to split part of the program into different files for reusablity.
Often used function can be saved in a module and imported into different programs

To make use of the functionality of any module, we need to import it into current program by using the import KW along
with the desired module name

The function, classes, variables etc can be accessed with the dot(.) operator along with the module name

Modules are also used to define python standard library
```
