```python
SET:
- a set is a collection which is unordered and unindexed
- sets are written within curly brackets {}
- It does not contain duplicate values. Set elements are unique
-sets are unordered, hence their elements appear at the output in any randome order
-a set itself may be modified, but the elements contained in the set must be of immutable type
```

```python
#eg:
s = {"Google", "Microsoft", "Apple", "IBM"}
print(s)
print(type(s))
```

```python
print(s[1])
```

```python
- We cannot access elements in a set by referring to an index, since sets are unordered and the items do not have any index
- we can loop through the set items using a loop
-we can use the "in" KW to check a particular value present in a set or not
```

```python
s = {"Google", "Microsoft", "Apple", "IBM"}
for i in s:
    print(i)
```

```python
# Check whether a value is present in the set or not
print("Apple" in s)
```

```python
print("Oracle" in s)
```

```python
print("Oracle" not in s)
```

```python
#Sets can have different datatypes
s1 = {1, 2, 3, 4, 5}
s2 = {1.2, 2.3, 3.4, 4.5, 5.5}
s3 = {(1, 2, 3), (1, 2), (8, 9, 10, 12)}
s4 = {5, "Python", 123.456, (10, 20, 30)}
print(s1)
print(s2)
print(s3)
print(s4)
print(f'{type(s1)}, {type(s2)}, {type(s3)}, {type(s4)}')
```

```python
SET METHODS:
add() : adds an elements to the set
clear() : removes all elements from a set
copy() : Returns a copy of a set
discard() : removes a specified element 
intersection(): returns a set, that is the intersection of 2 other sets
intersection_update() : Removes the items in the set that are not present in other specified set
isdisjoint() : returns whether 2 sets have intersection or not
issubset() : returns whether another set contains set or not
issuperset() : returns whether this set contains another set or not
pop(): removes and returns an elements from the set
remove(): removes a specified elements from the set
symmetric_difference() : Returns a set with symmetric differences of 2 sets
syymetric_difference_update(): Inserts the symmetric difference from this set and another
union(): returns a set containing the union of sets
```

```python
# to add an elements in to a set
s = {'Google', 'Microsoft', 'Apple', 'IBM'}
print(s)
s.add("Oracle")
print(s)
```

```python
#Update function is used to add multiple values from a tuple, set or a list
s = {'Google', 'Microsoft', 'Apple', 'IBM'}
print(s)
s.update(("oracle", "hp", "samsung"))
print(s)
```

```python
#set does not contain duplicate values
s = {'Google', 'Microsoft', 'Apple', 'IBM'}
print(s)
s.add("Apple")
print(s)
```

```python
# delete an element using remove() or discard()
#if the element are to be removed is not present in the set, remove() will raise an error
s = {'Google', 'Microsoft', 'Apple', 'IBM'}
print(s)
s.remove("Apple")
print(s)
```

```python
s = {'Google', 'Microsoft', 'Apple', 'IBM'}
print(s)
s.discard("Apple")
print(s)
s.discard("Apple")
print(s)
```

```python
#pop()
s.pop() - deletes an item. Usually deletes the last element and returns. However, as sets are unindexed 
# and unordered any random element is removed
s = {'Google', 'Microsoft', 'Apple', 'IBM'}
print(s)
x = s.pop()
print(x)
```

```python
s = {'Google', 'Microsoft', 'Apple', 'IBM'}
print(s)
x = s.clear()
print(x)
```

```python
isdisjoint() returns True is none of the items in both the sets are similar, otherwise returns False
a = {1, 2, 3, 4, 5, 6, 7}
b = {6, 7, 8, 9, 10,80}
print(a.isdisjoint(b))
```

```python
issubset() - returns True if all items in the set exists in the specified set, otherwise returns False
a = {1, 2, 3}
b = {10, 2, 12, 1, 15, 3, 5, 6}
print(a.issubset(b))
print(a.issubset(a))
```

```python
print(a.issuperset(b))
print(b.issuperset(a))
```
