```python
#length of each word
#Transforming 
wood = 'How much wood would a woodchunk chunk if a woodchunk could chunk wood?'.split()
# print all words having length >=5
x = [len(i) for i in wood]
print(x)
```

```python
#Append "-away" with each word that are less than 5 characters in length
wood = 'How much wood would a woodchunk chunk if a woodchunk could chunk wood?'.split()
a = [i+'-away' for i in wood if len(i)<5]
print(a)
```

```python
# Transfroming into tuple for each word - complex transforming
#filtering in words that are 5+ characters long and outputing the words with their length as tuples
wood = 'How much wood would a woodchunk chunk if a woodchunk could chunk wood?'.split()
a = [(i, len(i)) for i in wood if len(i)>=5]
```

```python
Dictionary Comprehension:
Similar to List comprehension. Syntax is:
out_dict = {key:value for (key,value) in iterable if condition}
```

```python
d1 = {'a':1, 'b':2, 'c':3, 'd':4}
#print all keys 
print(d1.keys())
#print all values
print(d1.values())
#print key value pairs
print(d1.items())
```

```python
#Accessing all values using traversals
d1 = {'a':1, 'b':2, 'c':3, 'd':4}
for k,v in d1.items():
    print(k,v)
```

```python
#Accesing all values using comprehension 
d1 = {'a':1, 'b':2, 'c':3, 'd':4}
d2 = {k:v for k,v in d1.items()}
print(d2)
```

```python
L1 = [1,2,3,4]
d = {i:i*i for i in L1}
print(d)
```

```python
L1 = [1,2,3,4]
d = {i:L1 for i in L1}
print(d)
```
