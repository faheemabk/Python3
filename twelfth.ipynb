```python
marks = {
    'Amir': {'English': 87, 'Maths': 92, 'Bio': 78},
    'Salman': {'English': 72, 'Maths': 64, 'Bio': 70},
    'Saif': {'English': 72, 'Maths': 64, 'Bio': 70}
}

print(marks)
```

```python
#print the english marks of saif
print(marks['Saif']['English'])
```

```python
Output:
Amir marks:
English => 87
Maths => 92
Bio => 78
Salman marks:
...
...
Saif marks:
...
```

```python
print('Amirs marks:')

print(marks['Amir'])
print('Salmans marks:')

print(marks['Salman'])
print('Saifs marks:')

print(marks['Saif'])
```

```python
for student in ['Amir', 'Salman', 'Saif']:
    print(f'{student} marks:')
    for subject, score in marks[student].items():
        print(f'{subject} => {score}')
```

```python
#Wap to count the number of characters occured in a given string 
s = 'Welcome to Python Programming'
d = {}
for i in s:
    d[i] = d.get(i, 0) + 1
for i,j in d.items():
    print(f'{i} -> {j}')
```

```python
#Wap to count the occurence of each word in a given string
s = 'New to python or choosing between python2 to python3? Learn Python2 or Python3?'
s = s.split()
d = {}
for i in s:
    d[i] = d.get(i, 0) + 1

for i,j in d.items():
    print(f'{i} -> {j}')
```

```python
# Convert 2 lists into a dictionary such that keys are from one list and values from list 2
k = ["Ten", "Twenty", 'Thirty']
v = [10, 20, 30]
d = {}
for i in range(len(k)):
    d[k[i]] = v[i]

print(d)
```

```python
#WAP to copy one dictionary followed by the second in a third
d1 = {'Ten':10, 'Twenty':20, 'Thirty':30}
d2 = {'Twenty':2020, 'Forty': 4040, 'Fifty':5050}

d3 = d1.copy()
print(d3)

d3.update(d2)
print(d3)
```
