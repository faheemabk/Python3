```python
# Multiple conditions
d = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5}
d1 = {k:v for k,v in d.items() if v>2 if v%2==0}
print(d1)
```

```python
# Three conditions
d = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5, 'f':6}
d1 = {k:v for k,v in d.items() if v>2 if v%2==0 if v%3==0}
print(d1)
```

```python
# Three conditions
d = {'a':1, 'b':2, 'c':3, 'd':4, 'e':5, 'f':6}
d1 = {k:v for k,v in d.items() if v>2 if v%2==0 if v%3==0}
print(d1)
```

```python
#Dictionary comprehension using enumerate function
L1 = [[73, -2345, 5], ['Land', 'Sea', 'Key']]
d = {i:j for i,j in enumerate(L1[1], 1)}
print(d)
```

```python
Set comprehension: are similar to list comprehension, however {} are used and only one single value is returned 
```

```python
# create an output set which can contain only even numbers from as list.
# Set removes duplicate values
L1 = [1,2,3,4,2,4,5,6,7,4,3,1]
s = {x for x in L1 if x%2==0}
print(s)
```
