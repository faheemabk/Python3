```python
f-string format : supports from version 3.6 onwards
f-string evaluates at run time of the program
It is swift compared to the previous and runs twice as fast
f-string has an easy syntax compare to other string format

syntax:
f"string"
every f-string statements consists of 2 parts.
One is character f or F and 
    the second is a string which needs to be formatted. This string is enclosed in quotes(single, double or triple)
```

```python
#eg:
print(f'Hello, welcome to python programming')
```

```python
name = 'sonu'
age = 21
print(f'Name is {name} and age is {age}')
```

```python
#f-string also supports use of functions
str = "welcome to python programming"
print(f'{str.title()}')
print(f'{str}')
```

```python
#f-string evaluates at run time, hence also supports matchematical expressions execution
print(f'{3**3}')
```

```python
x = 123.45678978
print(f'the value of x is {x}')
print(f'the value of x is {x:.3f}')
print(f'the value of x is {x:10.3f}')
print(f'the value of x is {x:<10.0f}')
print(f'the value of x is {x:^10.0f}')
print(f'the value of x is {x:*^10.0f}')
```

```python
for i in range(1, 5):
    print(f"{i:<6d} {i**2:<6d} {i**3:6d} {i**4:6d}")
```
