```python
Conditions are used for decision making in anticipation of conditions occuring during execution of a program
and specified actions taken according to the conditions

It always produces Boolean results : True and False as the outcome

Note: Python assumes any non-zero and not Null value as True and zero or Null as False
```

```python
Conditional Statements :
1. if statements - An if statement consists of a boolean expression followed by one or 
more statements
2. if.....else statements - an if statement can be followed by an else statement. It is optional 
It executes when an if statements is False
3. nested if statements - one or more if-else statements inside an if or else
```

```python
Syntax:
if condition:
    statement 1 
    statement 2
    statement 3
    .....
else :
    statement 1 
    statement 2
    statement 3
    .....
    
    if the condition is True, the if suite executes and the else suite skips 
    if the condition is False, the else suite is executed and the if suite is skipped
    the if statement can be used without the else
```

```python
#Example
val1 = 100
if val1 : 
    print("value 1 is true")
    print(val1)
    val2 = 0
if val2:
    print("value 2 is True")
    print(val2)
print("Done with the example")
```

```python
#example :
amt = int(input("enter a value"))

if amt<1000:
    disc = amt*0.5
elif amt<5000:
    disc = amt*0.15
else:
    disc = amt*0.15
print("discount is", disc)
print("net amount is ", amt-disc)
```

```python
# Nesting of if-else
if condtion:
    statement(s)
    if condition:
        statement(s)
    elif condition:
        statement(s)
    else:
        statement(s)
    elif condition:
        statement(s)
    else:
        statement(s)
        
```
