```python
# Number Type Conversion
1. int(x) => converts x into int
2. float(x) => converts x into float
3. complex(x, y) =>  converts into complex number with real part x and imaginary part 0 
4. complex(x, y)) => converts into complex with x into real and y into imaginary part
```

```python
a = 3.14
b = int(a)
c = float(b)
print(a, b, c)
```

```python
e = complex(b, c)
print(e)
```
