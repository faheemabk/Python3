```python
# Conditional Expressions: (Ternary Operators)
Syntax:
<exp1> if <condition> else <exp2>

returns exp1 if condition is True, otherwise returns exp2
```

```python
a = 10
b = 20
if a > b:
    m = a
else:
    m = b
print(m)
```

```python
# a certain steel is graded as per the following criteria
1. Hardness should be greater than 8900
2. Carbon contents should be less than 0.68
3. Tensile Strenght should be greater than 4550

The Grades Are as follows
1. grade is 10 if all 3 conditions are met
2. grade is 9 if condition 1 and 2 are met
3. grade is 8 condition if condition 2 and 3 are met
4. grade is 7 if condition 1 and 3 are met
5. gade is 6 if any one condition is met
6. grade is 5 if none of the conditions are met

if hardness of steel, carbon contents and tensile strength are input through the keyboard
write a program to print grade of steel
```

```python
# Accept 3 values from user and print the max value using conditional expression
a = 10
b = 50
c = 2000

m = a if (a > b and a > c) else (b if b > c else c)
print("max value is:", m)
```

```python
Iterative Statements : 
1. While loop - repeats statements or group of statements while a given condition is True.
2. for loop - Executes a sequence of statements multiple lines
3. Nested loop - one or more loop inside any loop
    
```

```python
while loop: 
    repeatedly executes a target statement as long as condition is True
    Syntax: 
    while condition:
          statements(s)
    else: 
          statements(s)
        
```

```python
Note: The else statement is optiona. when used inside while, it executes when the condition becomes false
```

```python
i = 1
while i<=5:
    print(i)
    i = i+1
print("Done with the loop")
```

```python
i = 1
while i<=100:
     print(i)
     i = i+1;
print("done")
```

```python
For Loop: it is used to iterate over a sequence (String, list, tuple, set etc)
syntax : 
for var in sequence :
else:
    statements
NOTE: var is a dummy variable defined locally for the loop.
the else clause is optional and is executed when the for loop is completed (condition becomes false)
```

```python
The range() function is used as an iterator. It returns a sequence of numbers, starting from 0 by default and increments by 1 default,
and stops before a specified number 
syntax:
range(start, stop, step)
Start is 0 by default, step is 1 by default. both are optional
```

```python
print(range(5))
```

```python
for i in range(5):
    print(i, end='                        ')
```

```python
for i in range(1, 101):
    print(i, end='    ')
```

```python
#print odd values from 1 to 100
for i in range(1, 101, 2):
    print(i, end='    ')
```

```python
for i in range(10,0, -1):
    print(i, end='    ')
```

```python
for i in range(1, 11):
    print("6 x ", i, "=", 6*1)
```
