```python
#Match-case : It is similar to switch-case. Supports 3.10 and above
i = 2
match i:
    case 1 :
        print("first case")
    case 2 :
        print("case two")
    case _ : #Default Case
        print("Default Case")
```

```python
We dont need to use break to prevent the code from continuing to the next case,
The default case is defined with the underscore, which will always match. We can combine several literals in a single pattern | ("or")
can also check with the if condition
```

```python
#Matchcase statement with the OR operator
n = int(input("enter number between 1-6"))

match n:
    case 1 | 2 :
        print("one or two")
    case 3 | 4 :
        print("three or four")
    case 5 | 6 :
        print("five or six")
    case _ :
        print("Number not between 1-6")
```

```python
# We can also use the python if condition along with match case statement when instead of 
# matching the exact value, we use a condition. Based on the condition, if the value is True 
#and matches the case pattern, its suite block is executed 

#Eg: Match case statement with python if condition 
n = int(input("Enter any number"))
match n:
    case n if n>0:
        print("positive")
    case n if n<0:
        print("negative")
    case _ :
        print("zero")
```

```python
fname='faheem'
lname='manyar'
age = 21
profession = 'learner'
course = 'python'
print("Hello %s %s, you are %d, you are a %s and you are learning %s"%(fname,lname,age,profession, course))
```

```python
# Option 2: str.format() - it is an improvement of %-formatting which allows multiple substitutions and value formatting.
# it makes use of of placeholders defined by {}
#eg:
fname='faheem'
lname='manyar'
age = 21
profession = 'learner'
course = 'python'
print("Hello {} {}, you are {}, you are a {} and you are learning {}".format(fname,lname,age,profession, course))
```

```python
# Errors occur when extra place holder is mentioned in the printing string - indexerror
print('{} is a boy'.format(lname))
```

```python
#When placeholders are empty, python replaces the values passed in the order as they
# appear in the str.format()
print('{} is {} years of age'.format(fname, age))
print('{0} is {1} years of age'.format(fname, age))
print('{1} is {0} years of age'.format(fname, age))
```

```python
print("the octal value of {:d} is {:o}".format(12,12))
print("the hexa value of {:d} is {:x}".format(12,12))
print("the HEXA value of {:d} is {:X}".format(12,12))
print("the Binary value of {:d} is {:b}".format(12,12))
```

```python
#String are left aligned by default and numbers right aligned
name = 'sonu'
age = 21
print("Name is {0:20s}, Age is {1:20d}".format(name, age))
print("Name is {0:<20s}, Age is {1:20d}".format(name, age))
print("Name is {0:>20s}, Age is {1:20d}".format(name, age))
print("Name is {0:^20s}, Age is {1:20d}".format(name, age))
```

```python
#alignment if data
< for left align
> for right align
^ for center align
name = 'sonu'
age = 21
print("Name is {0:20s}, Age is {1:20d}".format(name, age))
print("Name is {0:<20s}, Age is {1:<20d}".format(name, age))
print("Name is {0:>20s}, Age is {1:>20d}".format(name, age))
print("Name is {0:^20s}, Age is {1:^20d}".format(name, age))
```
