```python
#String functions
#capitalize(): capitalizes first character of a given string
str = "this is a string example....wow"
x = str.capitalize()
print(x)
```

```python
#center(width, fillchar) - Returns a string padded with fillchar with the original string centered
# to a total width columns
print(str.center(50))
print(str.center(50, '@'))
```

```python
# count(str, beg=0, end=len(string)) - counts number of times str occurs in a given string or sub-string
#starting from beg and ending at index end
str = "this is a string example....wow!!!!"
sub = 'i'
print(str.count(sub))
print(str.count(sub, 10, 40))
print(str.count('is'))
```

```python
#endswith(suffix, beg=0, end = len(string)) - checks if a given string ends with the suffix.
#returns true or false
str = 'this is a string example!!!'
s = '!!'
print(str.endswith(s))
print(str.endswith(s, 20))
print(str.endswith('exam', 0, 21))
```

```python
# exxpandtab(tabsize=8) - Expands tabs in a string to multiple spaces
str = "this is a \tstring example.....wow!!!!"
print('Hello ' + str)
print('Hello ' + str.expandtabs(8))
print('Hello ' + str.expandtabs(5))
```

```python
#find(str, beg=0, end=len(string)) : checks if a substring occurs in a string or not.
#Returns index if found, else returns-1
str="this is a string example...wow!!!"
substr="exam"
print(str.find(substr))
print(str.find('hello'))
```

```python
#rfind(str, beg=0, end=len(string)) : checks the last occurence of a substring in a string
#Returns index if found, else returns-1
str="this is a string example...wow!!!"
substr="is"
print(str.find(substr))
print(str.rfind(substr))
print(str.rfind('hello'))
```

```python
#index(substr, beg=0, end=len(string)) : same as find() but raises an exception if substr is not found
str = "this is a string example... wow!!!"
substr = "exam"
print(str.index(substr))
print(str.index('Hello'))
```

```python
# isalnum() : returns true if string has atleast 1 character and all characters are alphanumeric
# otherwise returns False. works with only single word strings
str = "Thisis2025"
print(str.isalnum())

str = "this is a string example...wow!!!"
print(str.isalnum())

str = "This is 2025"
print(str.isalnum())
```

```python
# isalpha() : returns true if string has atleast 1 character and all characters are alphabets
#otherwise returns False
str = "Thisis2025"
print(str.isalpha())

str = "this is a string example...wow!!!"
print(str.isalpha())

str = "ThisIsAstring"
print(str.isalpha())
```

```python
#isdigit(): returns True if string contains only digits, False otherwise
str = "1234567890"
print(str.isdigit())

str = "this is a string example...wow!!!"
print(str.isdigit())
```

```python
#islower() : Returns True is a string is lowercase, False otherwise
str = 'this is a string example...wow!!!'
print(str.islower())

str = 'This is a string example'
print(str.islower())
```

```python
#isspace() : returns True if string contains only whitespace character, returns False otherwise
str = '  '
print(str.isspace())

str = "this is a string example...wow!!!"
print(str.isspace())
```

```python
#istitle() : returns True if string is in title case, returns False otherwise
str = "THIS IS A STRING EXAMPLE WOW!!!"
print(str.isupper())

str = "This is a string example...wow!!!"
print(str.isupper())
```

```python
#join(seq) : merges (concatenates) a string representing elements in sequence into a string with
# a seperator string
s = '-'
seq = ('a', 'b', 'c')
print(s.join(seq))
```

```python
# len(string) - returns length of a given string
str = "this is a string example...wow!!!"
print(len(str))
```

```python
# ljust (width [, fillchar]) : returns a space padded string with the original string left justified
# to the total of width columns
str = "this is a string example...wow!!!"
print(str.ljust(50))
print(str.ljust(50, '*'))
```

```python
# lower(): converts a string into lowercase
str = "THIS IS A TEST RUN"
print(str.lower())

str = 'This is a python 3.13 envirOnment'
print(str.lower())
```

```python
# upper(): converts a string into lowercase
str = "THIS IS A TEST RUN"
print(str.upper())

str = 'This is a python 3.13 envirOnment'
print(str.upper())
```

```python
#lstrip() : removes all leading whitespaces in a string

str = "        this is a string example...!!!!!!"
print(str)
print(str.lstrip())

str = "\tthis is a string example...wow!!!"
print(str)
print(str.lstrip())

str = "******this is a string example.....wow!!!********"
print(str)
print(str.lstrip("*"))
```

```python
#lstrip() : removes all trailing whitespaces in a string

str = "            this is a string example...!!!!!!                    "
print(str + "Hello")
print(str.strip() + "Hello")

str = "\tthis is a string example...wow!!!\t"
print(str + "Hello")
print(str.strip() + "Hello")

str = "******this is a string example.....wow!!!********"
print(str + "Hello")
print(str.strip("*") + "Hello")
```

```python
#strip() : removes all leading and trailing whitespaces in a string

str = "     this is a string example...!!!!!!           "
print(str + "Hello")
print(str.strip() + "Hello")

str = "\tthis is a string example...wow!!!\t"
print(str + "Hello")
print(str.strip() + "Hello")

str = "******this is a string example.....wow!!!********"
print(str + "Hello")
print(str.strip("*") + "Hello")
```

```python
#max(str) : returns the maximum alphabet character from a string
str = "this is a string example...wow!!!"
print(max(str))

str = "this is a string demo...wow!!!"
print(max(str))
```

```python
#min(str) : returns the minimum alphabet character from a string
str = "thisisastringexamplewow"
print(min(str))

str = "thisisastringdemowow"
print(min(str))
```

```python
#replace(old, new[,max]) : replaces all occurences of old in string with new or atmost max occurence if 
#max is provided
str = "this is a string demo!!! This is really interesting"
print(str)
print(str.replace("is", "was"))
print(str.replace("  is", "  was", 2))
print(str.replace(" is", " was"))
```
