```python
Conversion in python: Convert datatypes. There are 2 types:
1.Implicit type conversion : python internally automatically converts data from one type to another type
whithout any user intervention.
2.Explicit type conversion: Python provides functions to convert data from one type to another
```

```python
#Implicit Data Conversion(Internal Conversion)
x1 = 10
print(f'Type of x1 is {type(x1)}')

x2 = 15
print(f'Type of x2 is {type(x2)}')

z1 = x1 + x2
print(f'Type of z1 is {type(z1)}')

f1 = 2.45
print(f'Type of f1 is {type(x1)}')

f2 = 3.1429
print(f'Type of f2 is {type(f2)}')

z2 = f1 + f2
print(f'Type of z2 is {type(z2)}')

z3 = x1 + f1
print(f'Type of z3 is {type(z3)}')

```

```python
#Python does not support internal conversion between strings and integers
i = 123
s = "345"
print(f'Type of i is {type(i)}')
print(f'Type of s is {type(s)}')
a = i + s 
print(f'Type of a is {type(a)}')
```

```python
In the above program, python is not able to convert string into integer. Hence in such situation, we make use of explicit
conversions using the type conversion functions
```

```python
#EG:
i = 123
s = "345"
print(f'Type of i is {type(i)}')
print(f'Type of s is {type(s)}')
a = i + int(s) #Explicitly converting string into integer type using the method int()
print(f'Type of a is {type(a)}')
```

```python
int() => converts value into integer value
Syntax : 
int(value, base)

The base is 10 by default. The second argument is optional 
```

```python
#int with floats
print(int(123.789))
print(int(23e5))
```

```python
# The int() function takes a string or float value as input and converts it into integer

print(f'3.14 => {int(3.14)}')
print(f'3.98 => {int(3.98)}')
print(f'3.99 => {int(-3.99)}')
print(f'3.14 => {int("2345")}')
print(f'3.14 => {int(14)}')

print(f'"123Hello" => {int("123Hello")}')
```

```python
# The int functions has a second argument which represents the base. The base is 10 bydefault(decimal)
#It can be changed to 2(binary), 8(octal) and 16(Hexadecimal)
s = '1100'
print(f's => {s}')
print(f'int(s, 2) => {int(s, 2)}')
print(f'int(s, 2) => {int(s, 8)}')
print(f'int(s, 2) => {int(s, 16)}')
print(f'int(s, 2) => {int(s, 10)}')
```

```python
print(int('10010', 2))
print(int(0b10010))

print(int('25', 8))
print(int(0o25))

print(int('A', 16))
print(int(0xA))
```

```python
#float() => converts number of string into float
print(f' int o float float(10)) => {float(10)}')
print(f' int o float float(-10)) => {float(-10)}')
print(f' int o float float(10)) => {float(123.456)}')
print(f' int o float float(10)) => {float("1234")}')
print(f' int o float float(10)) => {float("12.34")}')


#can also convert scientific notations
print(f'float(3e-002) => {float(3e-002)}')
print(f'float(3e-002) => {float("3e-002")}')
print(f'float(3e-002) => {float(+3E4)}')

#Also works with boolean, nan, infinity
print(f'True to float float(True) => {float(True)}')
print(f'True to float float(True) => {float(False)}')
print(f'True to float float(True) => {float("nan")}')
```
