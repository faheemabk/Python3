```python
Python identifiers : (Also referred to as names)
it is a name given to identify a variable, function, class, module, or other objects
It can be a combination of Alphabets, digits and underscore(_)
It should never start with a digit.
Case sensitive

Class names start with upper case (Title Case)
All identifiers are lower case 
Single leading underscore for a variable indicates variable is private
Two leading underscores indicate a strong private identifier
Identifier starting and enough with 2 underscores is defined as a special name
```
