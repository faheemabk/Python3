```python
# List slicing:
a[x:y:z]
where :
x is starting index which is included in the slicing
if x is not defined, slicing starts from 0 by default
y is the ending index which is excluded in the slicing 
if y is not defined, slicing ends at the end by default
z is the number of steps for increment or decrement of the slicing
if z is not defined, slicing is incremented by 1 by default
```

```python
#eg:
a = [0, 11, 22, 33, 44, 55, 66, 77, 88, 99]
print(a)
print(a[:])
print(a[::])
print(a[0::])
print(a[::1])
print(a[0::1])
print(a[:10:1])
print(a[0:10:1])
print(a[:10:1])
```

```python
print(a[0:10:2])
```

```python
#if z is not given, it is taken as 1 by default
a = [0, 11, 22, 33, 44, 55, 66, 77, 88, 99]
print(a[0:7:])
```

```python
# if z is positive, slicing will be from left to right, hence x<y
a = [0, 11, 22, 33, 44, 55, 66, 77 ,88, 99]
print(a[0:7:1])
print(a[7:2:1])
```

```python
# if z is negative, slicing will be from right to left, hence x>y
a = [0, 11, 22, 33, 44, 55, 66, 77 ,88, 99]
print(a[7:2:-1])
```

```python
#if z is positive and:
#if x is not given, x is taken as 0 by default
#if y is not given, y is taken as length of the list by default
a = [0, 11, 22, 33, 44, 55, 66, 77, 88, 99]
print(a[:7:1])
print(a[5::1])
print(a[::1])
```

```python
#if z is positive and:
#if x is not given, x is taken as -1 by default (index of the last element)
#if y is not given, y is taken as length of the list by default
a = [0, 11, 22, 33, 44, 55, 66, 77, 88, 99]
print(a[:-5:-1])
print(a[-5::-1])
print(a[::-1])
```

```python
Functions used with lists
1. len(list) -> returns the total length of a list
2. max(list) -> returns item from the list having max value
3. min(list) -> returns item from the list having min value
4. list(seq) -> converts a sequence into a list
5. list.append(object) -> Appends an object to a list
6. list.count(obj) -> returns count of the number of times obj occurs in the list
7. list.extend(seq) -> appends the contents of sequence
8. list.index(obj) -> returns lower index in list that object appears
9. list.insert(insert, obj) -> inserts object obj into list at offset index
10. list.pop(obj=list[-1]) -> removes and returns last object of obj from the list
11. list.remove(obj) -> removes objects of a list in place
12. list.reverse() - reverses object of a list in place
13. list.sort([func]) - sorts object of a list, use compare function if given
```
