```python
zip() : the purpose of zip function is to map similar index of multipe containers so that this can be used as a single entity

syntax: 
zip(*iterators)
where iterators can be lists, strings, dict, user defined objects.
returns a single iterator having mapped values from all the container
```

```python
#eg : 
names = ["Sonu", 'Monu', 'Golu', 'Chinku']
rollno = [4,2,3,1]
marks = [70, 65, 80, 78]

m = zip(names, rollno, marks)
print(m)
print(list(m))
```

```python
m = list(m)
```

```python
print(m)
```

```python
#unzip - unzipping means converting the zipped values back to the individual self as they were. This is done by the '*'
#operator along with the zip() function

names = ["Sonu", 'Monu', 'Golu', 'Chinku']
rollno = [4,2,3,1]
marks = [70, 65, 80, 78]

m = zip(names, rollno, marks)

n,r,m = zip(*m)
print(n)
print(r)
print(m)
```

```python
questions = ['NAME', 'QUEST', "FAVOURITE COLOUR"]
answers = ['GOLU', 'PROGRAMMING', 'BLUE']
for q,a in zip(questions, answers):
    print(f'What is you {q}? It is {a}')
    
```

```python
# Creating a dictionaries using zip function
names = ['Sonu', 'Monu', 'Golu', 'Chinku', 'Pinku']
marks = [75, 60, 80, 90, 85]
z1 = zip(names, marks)
d = {}

for i in z1:
    d[i[0]] = i[1]

print(d)
```

```python
#Using map to create zip
s = ['a', 'b', 'c', 'd', 'e']
n = [1, 2, 3, 4, 5]
results = list(zip(s, n))
print(results)
```

```python
#Same as
s = ['a', 'b', 'c', 'd', 'e']
n = [1,2,3,4,5]
results = list(map(lambda x,y :(x,y), s,n))
print(results)
```

```python
enumerate() function: it adds a counter to an iterable and returns it in a form of enumerate object
#which can directly be used in for loops or be converted into list or tuple
syntax:
enumerate(iterable, start=0)
```

```python
food = ['Bread', 'Milk', 'Butter']
ef = enumerate(food)
print(type(ef))
print(ef)
print(list(ef))
```

```python
food = ['Bread', 'Milk', 'Butter']
ef = enumerate(food, 101)
```

```python
food = ['Bread', 'Milk', 'Butter']
for i in enumerate(food):
    print(i)

print(i)

for i, j in enumerate(food, 501):
    print(f'{i} -> {j}')
```

```python
# enumerate with strings - can be used to show index of each character
name = 'Sonu Singh'
for i in enumerate(name):
    print(i)
```

```python
Comprehensions: they provide a short and concise way to construct new sequences such as list, sets, dictionary etc.
using sequences which have been already defined
Python supports 4 types of comprehensions :
1.List comprehension
2.Dictionary Comprehension
3.Set comprehension
4.Generator Comprehension
```

```python
List Comprehension : Provides an elegant way to create new lists
Syntax:
output_list = [output_exp for var in nput_list if(var satisfies this condition)]
or
list_var = [x for x in list]

List comprehensions may or may not contain an if condition
List comprehension can contain multiple nested list comprehensions
```

```python
# Eg:
L1 = [1, 2, 3, 4, 5]
L2 = [i for i in L1]
print(L2)
```

```python
Note: List comprehension is made inside the same bracket of a list []
[i for i in list] is the basic syntax where the first i is the element, second i is the iterator
and list is the iterable
```

```python
# Create a list which has elements thar are square of a given list
#Eg :
L1 = [1, 2, 3, 4, 5]
L2 = [i*i for i in L1]
print(L2)
```

```python
#using nested loop 
matrix = []
for i in range(5):
    matrix.append([])# Appending an empty list inside the list
    for j in range(5):
        matrix[i].append(j)
print(matrix)
```

```python
#Using nested list comprehension
matrix = [[j for j in range(5)] for i in range(5)]
print(matrix)
```
