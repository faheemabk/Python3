```python
# split(str="  ", num=string.count(str)) : splits a string according to delimiter str
# (and returns list of substrings)
str = "this is a string demo!!!"
print(str.split())
print(str.split('i'))
print(str.split('i', 2))
```

```python
#splitlines(num=string.count('\n')) : splits string at all new lines and returns a list of each line 
#with new lines removed
str = "this is \na string demo....\nwow!!!"
print(str)
```

```python
print(str.splitlines())
```

```python
#startswith(str, beg=0, end=len(string)) : checks if a strings starts with str.
#returns True is it does, otherwise False
str= "this is a string demo!!!"
print(str.startswith("this"))
print(str.startswith("string"))
print(str.startswith("string", 10))
```

```python
# swapcase() : inverses the case of a given string
str = "this is a demo for checking swapcase function!!!"
print(str)
print(str.swapcase())
```

```python
#title() : converts a string into titlecase
str = "this is a demo for 999 checking swapcase function!!"
print(str)
print(str.title())
```

```python
#zfill(width) : Returns original string left padded with zeros to a total width. Used with number
str = " this is a string demo!!! "
print(str.zfill(40))
```

```python
#isdecimal(): returns True is a string contains only decimal characters, False otherwise
str = "This2025"
print(str.isdecimal())

str = "20.25"
print(str.isdecimal())
```

```python
#isidentifier() : checks if a string is a valid python identifier or not. returns True if it is a 
#valid identifier, False otherwise
print('abc123'.isidentifier())
print('123abc'.isidentifier())
print('and'.isidentifier())
```

```python
# isprintable(): checks whether all characters in a string are printable or not.
str = 'this is a string demo...wow!!!'
print(str.isprintable())

str = 'this is a string \ndemo...wow!!!'
print(str.isprintable())
```

```python
#partition(<sep>) : splits a string on the first occurence of <sep>

str = "this is a string demo is.....wow!!!"
print(str.partition('is a'))
print(str.rpartition('is')) #splits on a last occurence
```
